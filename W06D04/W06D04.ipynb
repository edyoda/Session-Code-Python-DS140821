```python
class car:
    def ignition(self):
        self.speed = 0
        print('Our Car is On')
    def increaseSpeed(self):
        self.speed += 10
        print('Speed=',self.speed)
        
# speed is an attribute.
# ignition, increaseSpeed are methods.
```

```python
breeza = car()
```

```python
breeza.ignition()
```

```python
breeza.increaseSpeed()
```
