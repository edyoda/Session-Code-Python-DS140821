```python
What are api's
Application Programming Interface.

Api's are generally used to exchange data. (JSON is mostly used to 
interchange the data with the api's)

Consider that you are in a restraunt.
1) Call the waiter
2) you will tell him the order details (like pizza, cold üç∫  etc)
3) waiter will go back in the kitchen and tell the chef to make a pizza
and prepare cold drinks
4) waiter will come back after some time with prepared food.

API'S are analogous to a waiter. API == WAITER

1) WE WILL FORM THE API
2) WE WILL TELL THE API THE DATA THAT WE WANT, API WILL GO BACK TO DATABASE
AND TELL WHAT IT NEEDS.
3) API WILL COME BACK WITH THE RESPONSE
```

```python
NORMAL WEBSITES LIKE FACEBOOK VS API'S

NORMAL WEBSITES ARE FOR HUMAN READABILITY AND ACCESSIBILITY.
BUT API'S ARE USED BY CODE TO READ AND ACCESS THE DATA

API'S WILL GIVE DATA IN JSON, OR OTHER CODE READABLE FORMAT.(LIKE XML)
NORMAL WEBSITES GIVE DATA IN HTML OR HUMAN accessible FORMAT.
```

```python
EXAMPLE OF API'S 

UBER APP: IT USES API'S OF GOOGLE MAPS TO CALCULATE THE FARE FOR THE USER,
    AND OTHER THINGS
    
ZOMATO APP,ETC

API'S ARE PROPERLY REGISTERED AS WHO CAN USE THEM.
```

```python
TYPES OF API'S
GET - WE WANT SOME DATA FROM A SERVER

POST - WE WANT TO SEND SOME DATA TO A SERVER/WEBSITE.
```

# POST REQUESTS IN API'S

```python
import requests

url = "https://httpbin.org/post"

payload = "{'python':3,'java':8}"
headers = {
  'Content-Type': 'text/plain'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```
